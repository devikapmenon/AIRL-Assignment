```python
import os
import io
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import numpy as np
import torch
import cv2
def show_image(img, title=None, figsize=(10,6)):
    if isinstance(img, torch.Tensor):
        img = img.detach().cpu().numpy()
        if img.ndim==3:
            img = np.transpose(img, (1,2,0))
    if isinstance(img, np.ndarray):
        if img.dtype == np.float32 or img.dtype==np.float64:
            img = (np.clip(img,0,1)*255).astype(np.uint8)
        img = Image.fromarray(img)
    plt.figure(figsize=figsize)
    plt.axis('off')
    if title: plt.title(title)
    plt.imshow(img)
    plt.show()
def overlay_mask_on_image(image:Image.Image, mask:np.ndarray, color=(255,0,0), alpha=0.4):
    image = image.convert('RGBA')
    mask_img = Image.new('RGBA', image.size, (0,0,0,0))
    mask_arr = (mask>0).astype(np.uint8)*255
    mask_pil = Image.fromarray(mask_arr, mode='L')
    color_img = Image.new('RGBA', image.size, color + (0,))
    mask_img.paste(color_img, (0,0), mask_pil)
    blended = Image.alpha_composite(image, Image.new('RGBA', image.size))
    blended = Image.alpha_composite(image, mask_img)
    return blended
def download_if_needed(url, dst_path):
    if os.path.exists(dst_path):
        print(f"Found {dst_path}")
        return dst_path
    print(f"Downloading {url} -> {dst_path} (please provide correct URL in the notebook)")
    # In Colab you can use wget or gdown depending on host
    return dst_path
try:
    from groundingdino.models import build_model as build_gd_model
    from groundingdino.util.inference import load_model_and_predict
    groundingdino_available = True
except Exception as e:
    print('GroundingDINO not available in this runtime:', e)
    groundingdino_available = False

# Fallback: CLIPSeg zero-shot which returns masks directly from text (no boxes), good for simple cases.
try:
    from clipseg import CLIPDensePredT
    clipseg_available = True
except Exception as e:
    clipseg_available = False
def get_boxes_from_groundingdino(image_pil:Image.Image, prompt:str, box_threshold=0.3):
    """Uses GroundingDINO to return list of boxes in xyxy normalized pixel coords and scores/labels.
    Replace with the exact call from the GroundingDINO repo you've installed."""
    if not groundingdino_available:
        raise RuntimeError('GroundingDINO not installed in the environment')
    # --- PSEUDO implementation outline ---
    # 1. preprocess image
    # 2. run model with (image, prompt)
    # 3. parse outputs -> boxes (x0,y0,x1,y1) in image pixel coords
    boxes = []
    scores = []
    labels = []
    # TODO: replace with repo-specific code
    return boxes, scores, labels

# CLIPSeg pathway (simple): returns a floating mask the size of the image

def get_mask_from_clipseg(image_pil:Image.Image, prompt:str):
    if not clipseg_available:
        raise RuntimeError('CLIPSeg not installed')
    model = CLIPDensePredT(version="default")
    model.eval()
    transform = None
    # CLIPSeg returns logits — typical pipeline uses resizing and normalization
    # This is a placeholder; follow CLIPSeg examples for exact preprocessing
    img = np.array(image_pil)
    # TODO: perform transforms
    # mask = model.predict(img, prompt)
    mask = np.zeros((image_pil.size[1], image_pil.size[0]), dtype=np.uint8)
    return mask
try:
    from sam2 import SamPredictor, sam_model_registry
    sam_available = True
except Exception as e:
    print('SAM2 import failed (may need to install from repo):', e)
    sam_available = False


def run_sam_with_boxes(image_pil:Image.Image, boxes:list):
    """boxes: list of [x0,y0,x1,y1] in pixel coords. Returns list of masks (H,W) boolean.
    Implementation depends on SAM2 API; this is skeleton code."""
    if not sam_available:
        raise RuntimeError('SAM2 not available')
    # Example (pseudo):
    # model = sam_model_registry['sam_vit_h'](checkpoint='path/to/checkpoint')
    # predictor = SamPredictor(model)
    # predictor.set_image(np.array(image_pil))
    # transformed_boxes = convert_pixel_to_sam_input(boxes)
    # masks, scores, logits = predictor.predict_boxes(transformed_boxes)
    masks = []
    return masks
def segment_with_text(image_pil:Image.Image, prompt:str, method='groundingdino'):
    """High-level pipeline:
    - text -> boxes (GroundingDINO) OR mask (CLIPSeg)
    - boxes -> SAM2 prompt -> masks
    - return masks and overlay
    """
    if method=='groundingdino' and groundingdino_available:
        boxes, scores, labels = get_boxes_from_groundingdino(image_pil, prompt)
        if len(boxes)==0:
            print('No boxes found with GroundingDINO for prompt:', prompt)
            return None
        masks = run_sam_with_boxes(image_pil, boxes)
        # For demo, pick the highest score mask
        if len(masks)>0:
            return masks[0]
        return None
    elif method=='clipseg' and clipseg_available:
        mask = get_mask_from_clipseg(image_pil, prompt)
        return mask
    else:
        raise RuntimeError('No valid method available. Please install GroundingDINO or CLIPSeg.')
if __name__ == '__main__':
    # Demo image - in Colab replace with upload UI or a URL
    demo_path = '/content/demo.jpg'  # replace with your image path
    if not os.path.exists(demo_path):
        print('Place an image at', demo_path, 'or modify the path to your image.')
    else:
        img = Image.open(demo_path).convert('RGB')
        show_image(img, 'Input image')
        prompt = 'Butterfly'  # replace with your text prompt
        try:
            mask = segment_with_text(img, prompt, method='groundingdino')
            if mask is None:
                print('No mask returned by the pipeline')
            else:
                blended = overlay_mask_on_image(img, mask)
                show_image(blended, 'Segmentation overlay')
        except Exception as e:
            print('Pipeline failed:', e)

def propagate_mask_with_optical_flow(frame0, frame1, mask0):
    """Warp mask0 from frame0 to frame1 using Farneback optical flow (grayscale)."""
    prev = cv2.cvtColor(np.array(frame0), cv2.COLOR_RGB2GRAY)
    nextf = cv2.cvtColor(np.array(frame1), cv2.COLOR_RGB2GRAY)
    flow = cv2.calcOpticalFlowFarneback(prev, nextf, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    h,w = mask0.shape
    grid_x, grid_y = np.meshgrid(np.arange(w), np.arange(h))
    map_x = (grid_x + flow[...,0]).astype(np.float32)
    map_y = (grid_y + flow[...,1]).astype(np.float32)
    warped = cv2.remap(mask0.astype(np.uint8)*255, map_x, map_y, interpolation=cv2.INTER_LINEAR)
    return (warped>127).astype(np.uint8)

# Example video propagation pipeline (user provides frames list)

def propagate_over_video(frames:list, init_mask, init_frame_idx=0):
    masks = [None]*len(frames)
    masks[init_frame_idx] = init_mask
    # propagate forward
    for i in range(init_frame_idx, len(frames)-1):
        masks[i+1] = propagate_mask_with_optical_flow(frames[i], frames[i+1], masks[i])
    # propagate backward
    for i in range(init_frame_idx, 0, -1):
        masks[i-1] = propagate_mask_with_optical_flow(frames[i], frames[i-1], masks[i])
    return masks
```
